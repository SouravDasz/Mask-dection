```python
import cv2
import numpy
import torch
```

```python
from  torchvision.transforms import transforms
```

```python
train_transform=transforms.Compose([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5))
])
```

```python
from torchvision.datasets import ImageFolder

from torch.utils.data import DataLoader
```

```python
train_ds=ImageFolder('/content/data',transform=train_transform)
```

```python
print(train_ds.classes)
print(train_ds.class_to_idx)
```

```python
train_loader = DataLoader(train_ds,
                          batch_size=32,
                          shuffle=True)

```

```python
images, labels = next(iter(train_loader))

print(images.shape)  # (batch, channels, height, width)
print(labels.shape)
```

```python
import torch.nn as nn

class Mycnn(nn.Module):
    def __init__(self):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=2, padding=1),  # fixed padding
            nn.ReLU(),
            nn.BatchNorm2d(32),
            nn.MaxPool2d(2,2),

            nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1), # fixed
            nn.ReLU(),
            nn.MaxPool2d(2,2),

            nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1), # fixed
            nn.ReLU(),
            nn.MaxPool2d(2,2),
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(1152, 256),
            nn.ReLU(),
            nn.Linear(256, 1)   # single output for binary
            # ❌ removed Sigmoid
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = Mycnn().to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 10

for epoch in range(num_epochs):
    model.train()

    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:

        images = images.to(device)
        labels = labels.float().unsqueeze(1).to(device)
        # BCE expects float and shape (batch,1)

        # Forward
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

        # Accuracy calculation
        probs = torch.sigmoid(outputs)
        predicted = (probs > 0.5).float()

        correct += (predicted == labels).sum().item()
        total += labels.size(0)

    epoch_loss = running_loss / len(train_loader)
    epoch_acc = correct / total

    print(f"Epoch [{epoch+1}/{num_epochs}] "
          f"Loss: {epoch_loss:.4f} "
          f"Accuracy: {epoch_acc:.4f}")
```

```python
from PIL import Image
import torchvision.transforms as transforms
import torch

# 1️⃣ Set model to evaluation mode
model.eval()

# 2️⃣ Load image
img = Image.open(img_path).convert("RGB")

# 3️⃣ Apply same transform as training
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

img = transform(img)

# 4️⃣ Add batch dimension
img = img.unsqueeze(0).to(device)

# 5️⃣ Turn off gradient calculation
with torch.no_grad():
    output = model(img)          # raw number (logit)
    prob = torch.sigmoid(output) # convert to probability
    pred = (prob > 0.5).float()  # convert to 0 or 1

print("Probability:", prob.item())
print("Prediction:", int(pred.item()))
```
